```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{} %'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

objects = ('Radix sort', 'Binary search','Reduce', 'etc')
y_pos = np.arange(len(objects))
performance = [80,60,30,20]
colors = ['lightcoral','deepskyblue','pink','limegreen']

bar = plt.bar(y_pos, performance, align='center', color = colors)
autolabel(bar)
plt.xticks(y_pos, objects)
plt.ylabel('SM utilization in %')
plt.title('Kernel type')

plt.savefig('SM_perforance.pdf')
```

```python
import matplotlib.pyplot as plt
import numpy as np



# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Memcpy', 'Radix sort', 'Binary search', 'Others'
colors = ['lightcoral','deepskyblue','pink','limegreen']
sizes = [42, 32, 13, 2]
explode = (0.1, 0.1, 0.1, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', colors=colors,
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set(title='Kernels runtime in %')

plt.savefig('runtime.pdf')
```
